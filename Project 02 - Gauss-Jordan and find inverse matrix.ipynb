Nhìn chung thuật toán Gauss Jordan cũng giống với thuật toán khử Gauss tuy nhiên chỉ khác một số chỗ nhỏ. Ta có thể biểu diễn thuật toán như sau:
- Bước 1: Cho $i = 0$; $j = 0$
- Bước 2: Nếu $i > n-1$ và $j > m-1$ thì kết thúc.
- Bước 3: 
    - Nếu $a_{ij}\neq 0$, thực hiện các phép biến đổi sau:
        - $\frac{1}{a_{ij}}d_i$
        - $d_k - a_{ik}d_i ~ \forall k\neq i$
        
    Sau đó $i=i+1; j=j+1$ và quay về Bước 2
    - Nếu $a_{ij} = 0$ thì sang Bước 4.
- Bước 4:
    - Nếu $a_{kj} \neq 0$ với một $k>i$ nào đó thì chọn một $k$ như vậy và thực hiện phép biến đổi $d_i \leftrightarrow d_k$ và quay về Bước 3.
    - Nếu $a_{kj} = 0$ với mọi $k>i$ thì $j = j + 1$ và quay về Bước 2.
